## 2. Feature Engineering

### 2.1 Add a new column named timeofday to give insight of sales in the Morning, Afternoon and Evening. This will help answer the question on which part of the day most sales are made.

```python
#Add a new column named timeofday to give insight of sales in the Morning, Afternoon and Evening. This will help answer the question on which part of the day most sales are made.


query2="""select Date,
            round(sum(Morning),2) as Morning,
            round(sum(Afternoon),2) as Afternoon,
            round(sum(Evening),2) as Evening,
            round(sum(Night),2) as Night,
            round(SUM(Morning + Afternoon + Evening + Night),2) AS TotalDailySales 
from
    (select Date,
    case when extract(hour from Time) >=6 and extract(hour from Time) <12 then total else 0 end as Morning,
    case when extract(hour from Time) >=12 and extract(hour from Time) <16 then total else 0 end as AfterNoon,
    case when extract(hour from Time) >=16 and extract(hour from Time) <19 then total else 0 end as Evening, 
    case when extract(hour from Time) >=19 and extract(hour from Time) <=24 then total else 0 end as Night from amazon
            ) as subquery 
        group by Date 
        order by Date;"""
pd.read_sql(query2,connection)
```

###  2.2 Add a new column named dayname that contains the extracted days of the week on which the given transaction took place (Mon, Tue, Wed, Thur, Fri).

```python
# Add a new column named dayname that contains the extracted days of the week on which the given transaction took place (Mon, Tue, Wed, Thur, Fri). 
# timeofday 
query3="""select Date,CASE when DAYNAME(date)='Monday' THEN 'Mon' 
                           when DAYNAME(date)='Tuesday' THEN 'Tue' 
                           when DAYNAME(date)='Wednesday' THEN 'Wed' 
                           when DAYNAME(date)='Thursday' THEN 'Thu' 
                           when DAYNAME(date)='Friday' THEN 'Fri' 
                           when DAYNAME(date)='Saturday' THEN 'Sat' 
                           when DAYNAME(date)='Sunday' THEN 'Sun' 
 end as DayofWeek, round(sum(Total),2) as Total_amount, 
round(sum(Morning),2) as Morning, round(sum(Afternoon),2) as Afternoon, 
round(sum(Evening),2) as Evening, 
round(sum(Night),2) as Night 
    from 
        (select Date,Total,
case when extract(hour from Time) >=6 and extract(hour from Time) <12 then total else 0 end as Morning, 
case when extract(hour from Time) >=12 and extract(hour from Time) <16 then total else 0 end as AfterNoon, 
case when extract(hour from Time) >=16 and extract(hour from Time) <19 then total else 0 end as Evening, 
case when extract(hour from Time) >=19 and extract(hour from Time) <=24 then total else 0 end as Night from amazon
                    ) as subquery 
WHERE DAYNAME(date) IN ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday') 
group by Date order by Date;"""
pd.read_sql(query3,connection)
```

###  2.3 Add a new column named monthname that contains the extracted months of the year on which the given transaction took place (Jan, Feb, Mar).

```python
# Add a new column named monthname that contains the extracted months of the year on which the given transaction took place (Jan, Feb, Mar).
# dayname 
query6="""select extract(YEAR from Date) as Year,DATE_FORMAT(Date,'%b') AS Month,Date, 
case WHEN DAYNAME(date) = 'Monday' THEN 'Mon' 
     WHEN DAYNAME(date) = 'Tuesday' THEN 'Tue' 
     WHEN DAYNAME(date) = 'Wednesday' THEN 'Wed' 
     WHEN DAYNAME(date) = 'Thursday' THEN 'Thu' 
     WHEN DAYNAME(date) = 'Friday' THEN 'Fri' 
     WHEN DAYNAME(date) = 'Saturday' THEN 'Sat' 
     WHEN DAYNAME(date) = 'Sunday' THEN 'Sun' 
end as DayofWeek, round(sum(Total),2) as amount from amazon 
where DAYNAME(date) in ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday') 
group by Month,Date order by Date;"""
pd.read_sql(query6,connection)
```

```python
# Add a new column named monthname that contains the extracted months of the year on which the given transaction took place (Jan, Feb, Mar)
# monthname wise sale

query4=""" select extract(YEAR from Date) AS Year, DATE_FORMAT(Date, '%M') AS Month, ROUND(SUM(Total), 2) AS Total_Amount 
from amazon 
group by Year, Month 
UNION ALL 
select 'Total' AS Year, '' AS Month, round(SUM(Total), 2) AS Total_Amount 
from 
( select extract(YEAR from Date) AS Year, DATE_FORMAT(Date, '%M') AS Month, 
ROUND(SUM(Total), 2) AS Total from amazon GROUP BY Year, Month ) AS subquery 
order by Year, Month;"""
pd.read_sql(query4,connection)
```

```python
# Find out which product line is selling the most in the month of January

query5=""" select product_line, SUM(unit_price * Quantity) as Total_cost from amazon 
where MONTH(Date) = 1 
group by product_line 
order by total_cost DESC 
limit 1;"""
pd.read_sql(query5,connection)
```

```python
# find out total sales and quantity through each and every product_line in the month of february

query7=""" select product_line, SUM(quantity) as Total_sales, SUM(unit_price * quantity) as Total_Amount from amazon 
where MONTH(Date) = 2 
group by product_line;"""
pd.read_sql(query7,connection)
```

```python
# find out total sales and quantity through each and every product_line in the month of march

query8=""" select product_line, SUM(quantity) as Total_sale_quantity, SUM(unit_price * quantity) as Total_Amount from amazon 
where MONTH(Date) = 3 
group by product_line;"""
pd.read_sql(query8,connection)
```

```python
# find the TOTAL(quantity,tax unit_price and total sale values) for each month and year.

query9=""" select extract(YEAR from Date) as Year, DATE_FORMAT(Date, '%b') as Month, SUM(quantity) as Total_Sales_Quantity, 
            SUM(unit_price * quantity) as Total_Sale_Amount, ROUND(SUM(`Tax 5%`), 2) as Tax_Value, 
            ROUND(SUM(total), 2) as Total_Sale_Value from amazon 
        group by Year, Month 
UNION ALL
 select 'Total' as Year, '' as Month, SUM(Total_Sales_Quantity) as Total_Quantity, 
 ROUND(SUM(Total_Sale_Amount), 2) as Total_Amount, 
 SUM(Tax_Value) as Tax_Value, SUM(Total_Sale_Value) as Total_Value 
    from 
    ( select extract(YEAR from Date) as Year, 
DATE_FORMAT(Date, '%b') as Month, SUM(quantity) as Total_Sales_Quantity, SUM(unit_price * quantity) as Total_Sale_Amount, 
ROUND(SUM(`Tax 5%`), 2) as Tax_Value, ROUND(SUM(total), 2) as Total_Sale_Value from amazon 
group by Year, Month) as subquery;"""
pd.read_sql(query9,connection)
```

```python
# find the sales of sundays in january month

query10=""" select EXTRACT(YEAR FROM Date) as Year, DATE_FORMAT(Date, '%b') as Month, Date, 'Sun' as DayofWeek,
   ROUND(SUM(Total), 2) as amount 
from amazon 
where DAYOFWEEK(Date) = 1 and MONTH(Date) = 1
group by Month, Date 
order by Date; """
pd.read_sql(query10,connection)
```

```python
# find the sales of day by day in january month

query11=""" select EXTRACT(YEAR FROM Date) as Year, DATE_FORMAT(Date, '%b') as Month, Date,
   ROUND(SUM(Total), 2) as amount 
from amazon 
where MONTH(Date) = 1
group by Month, Date 
order by Date; """
pd.read_sql(query11,connection)
```

```python
# find the which month highest sales in year

query12=""" select extract(year FROM Date) AS Year,
CASE 
   when SUM(CASE WHEN MONTH(Date) = 1 THEN Total ELSE 0 END) > SUM(CASE WHEN MONTH(Date) = 2 THEN Total ELSE 0 END) AND 
        SUM(CASE WHEN MONTH(Date) = 1 THEN Total ELSE 0 END) > SUM(CASE WHEN MONTH(Date) = 3 THEN Total ELSE 0 END) 
        then 'January sales Highest'
   when SUM(CASE WHEN MONTH(Date) = 2 THEN Total ELSE 0 END) > SUM(CASE WHEN MONTH(Date) = 1 THEN Total ELSE 0 END) AND 
        SUM(CASE WHEN MONTH(Date) = 2 THEN Total ELSE 0 END) > SUM(CASE WHEN MONTH(Date) = 3 THEN Total ELSE 0 END) 
        then 'February sales Highest'
   when SUM(CASE WHEN MONTH(Date) = 3 THEN Total ELSE 0 END) > SUM(CASE WHEN MONTH(Date) = 1 THEN Total ELSE 0 END) AND 
        SUM(CASE WHEN MONTH(Date) = 3 THEN Total ELSE 0 END) > SUM(CASE WHEN MONTH(Date) = 2 THEN Total ELSE 0 END) 
        then 'March sales Highest'
        else 'Equal sales among months' end as Sales_Comparison
from amazon 
group by Year; """
pd.read_sql(query12,connection)
```

```python
# which city has the highest sales in every month

query13=""" with MonthlySales as (
    select extract(YEAR FROM Date) as Year, extract(MONTH FROM Date) as Month, City, round(SUM(Total),2) as Total_Sales,
        RANK() OVER(PARTITION BY EXTRACT(YEAR FROM Date), EXTRACT(MONTH FROM Date) ORDER BY SUM(Total) DESC) as rank_month
    from amazon
    group by extract(YEAR FROM Date), extract(MONTH FROM Date), City
) select Year, 
    CASE Month
        when 1 then 'Jan'
        when 2 then 'Feb'
        else 'Mar' end as Month, City, SUM(Total_Sales) OVER(PARTITION BY Year, Month) as Total_Sales_Month
from MonthlySales
where rank_month = 3;"""
pd.read_sql(query13,connection)
```

```python
# the weekly sale report for every month 

query14=""" select  
    extract(YEAR FROM Date) as Year,
    CONCAT('Week ', WEEK(Date)) as Week,
    DATE_FORMAT(MIN(Date), '%b %d') as Start_Date,
    DATE_FORMAT(MAX(Date), '%b %d') as End_Date,
    ROUND(SUM(Total), 2) as Amount
from amazon 
where MONTH(Date) = 1
group by Year, Week
order by MIN(Date); """
pd.read_sql(query14,connection)
```
